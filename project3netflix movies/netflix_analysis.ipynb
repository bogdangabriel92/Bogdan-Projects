```python
# Investigating Netflix Movies (1990s)
# Author: Bogdan-Gabriel Rasu
# Description: Exploratory Data Analysis on Netflix movies from the 1990s

# === 1. Importing Libraries ===
import pandas as pd
import matplotlib.pyplot as plt

# === 2. Loading Dataset ===
netflix_df = pd.read_csv("netflix_data.csv")

# Preview the first rows
print("Initial dataset preview:")
print(netflix_df.head())

# === 3. Filtering only 'Movie' type ===
netflix_movies = netflix_df[netflix_df["type"] == "Movie"]
print(f"\nTotal movies in dataset: {len(netflix_movies)}")

# === 4. Filtering movies from the 1990s ===
movies_1990s = netflix_movies[
    (netflix_movies["release_year"] >= 1990) &
    (netflix_movies["release_year"] < 2000)
]

print(f"Movies from the 1990s: {len(movies_1990s)}")

# === 5. Visualizing durations with a histogram ===
plt.figure(figsize=(10,6))
plt.hist(movies_1990s["duration"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Movie Durations (1990s)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.grid(True)
plt.show()

# === 6. Determine most frequent duration ===
# Option 1 (visual): choose manually based on histogram
duration = 100   # approximate value from histogram

# Option 2 (accurate):
# duration = movies_1990s["duration"].mode()[0]

print(f"Most common duration in the 1990s: {duration} minutes")

# === 7. Counting short Action movies (<90 min) ===
action_movies_1990s = movies_1990s[movies_1990s["genre"] == "Action"]

# Counting how many have duration < 90
short_movie_count = (action_movies_1990s["duration"] < 90).sum()

print(f"Number of short Action movies (<90min) in the 1990s: {short_movie_count}")

# === 8. Saving clean subset to CSV (optional) ===
movies_1990s.to_csv("netflix_movies_1990s.csv", index=False)
```
